La masa de 0.31 kg se desliza sobre un alambre sin fricción que está en el plano vertical. El resorte ideal unido a la masa tiene una longitud libre de 80 mm y su rigidez es 120 N/m.<dt>
$a.$ Calcule el menor valor de la distancia b si la masa debe alcanzar el extremo del alambre en B después de ser liberada, a partir del reposo, en A.<dt> $b.$ Tomando el resultado del literal a, encuentre el trabajo realizado por la fuerza elástica del punto A hasta el punto 3.

# $Resolución:$

PARTE A
>Realizaremos el DCL<dt>
![title](10.jpeg) <dt>
>Aplicando el principo de trabajo y energia tenemos:
$$T_1+U_{1-2}=T_2$$ <dt> $$\frac{1}{2}m{{v_1}^2}+U_{1-2}=\frac{1}{2}m{v_{2}^2}$$ <dt> Dado que las velocidades en el punto 1 y en el punto 2 son iguales cero, entonces:<dt> $$U_{1-2}=0$$ <dt> $$U_2-U_1=0$$ <dt> $$U_2=U_1$$ <dt> Ahora:<dt> $$-wh-\frac{1}{2}k({x_2}^2-{x_1}^2)=0$$ <dt> $$wh=-\frac{1}{2}k({x_2}^2-{x_1}^2)$$ <dt> Dado que en el punto 2 no existe deformacion en le resorte entonces $x_2$ es igual a cero. <dt> $$wh=\frac{1}{2}k{x_1}^2$$<dt> $$mg(h_2-h_1)=\frac{1}{2}k{x_1}^2$$ <dt> Sea $h_2-h_1=2R$ <dt> $$mg2R=\frac{1}{2}k{x_1}^2\qquad{(a)} $$ <dt> Para determinar $x_1$ nos basamos en el siguiente triangulo: <dt> ![title](2.png) $${L}^2={b^2}+{R^2}$$ <dt> Calculado ${x_1}$ <dt> $$x_1=L-L_o$$ <dt> $$L_o=R$$ <dt> $$L=\sqrt{b^2+R^2}$$ <dt> $$x_1=\sqrt{b^2+R^2}-R$$ <dt> Volviendo a la ecuacion (a)<dt> $$mg2R=\frac{1}{2}k\left(\sqrt{b^2+R^2}-R\right)^2$$ <dt> Reemplazando los datos: <dt> $$(0.31)(9.81)(2)(0.08)=\frac{1}{2}(120)\left(\sqrt{b^2+0.08^2}-0.08\right)^2$$ <dt> $$\left(\sqrt{b^2+0.08^2}-0.08\right)=0.008110$$ <dt> $$b=0,1501 [m]$$ <dt> $$b=150.1 [mm]$$



PARTE B

>Para esta parte del ejercicio tomaremos nuestro sistema de referecia de la siguiente manera:
![title](3.png)
>Sabemos que $$F_e=Kx$$ <dt> Como x varia acorde a b entonces: <dt> $$(x+R)^2=b^2+R^2$$ <dt> Donde: $$x=\sqrt{b^2+R^2}-R$$ <dt> Entonces: $$F_e=K(\sqrt{b^2+R^2}-R)$$ <dt> Entonces para el trabajo de la fuerza elastica tenemos: $$U_{1-2}=\frac{1}{2}k(\sqrt{b^2+R^2}-R)^2$$ <dt> tenemos que $$x=b$$ <dt> Tendríamos como paso final el trabajo vs la posicción $$U_{1-2}=\frac{1}{2}k(\sqrt{x^2+R^2}-R)^2$$



```python
#datos
b=np.sqrt((np.sqrt((4*0.31*9.81*0.08)/(120))+0.08)**2-(0.08**2))
#análisis
th=np.arange(-0.0001,b+0.0001,0.0001) #creando lista numérica
vs=60*((np.sqrt(th**2+0.08**2))-0.08)**2
#plot
fig=px.line(x=th,y=vs,labels={'x':'Distancia[m]','y':'Trabajo[J]'},title="TRABAJO DEL RESORTE: La deformación del resorte depende la distancia de b")
fig.show()
```
